```python
from sklearn.inspection import permutation_importance
import time 

from Script import *

df_train = pd.read_csv("Data/train.csv")
df_test  = pd.read_csv("Data/test.csv")

df_train = df_train.drop("i", axis=1)
y = df_train.pop("y")
df_train.insert(54,"y", y)
df_test = df_test.drop(["y", "i"], axis = 1)


df_train, df_test = preprocess_data(df_train, df_test, verbose=False)


X = df_train.iloc[:, 0:-1]
y = df_train.iloc[:, -1]

```

```python
def create_encoder(X):
    X_2 = X.select_dtypes(include=['category'])
    enc = OneHotEncoder(drop='first', handle_unknown='ignore')
    enc.fit(X_2)
    return enc

_x_encoder = create_encoder(X)

def prepare(X):
    cat_col = X.select_dtypes(include=['category']).columns
    aux_transformed = _x_encoder.transform(X[cat_col]).toarray()
    X.drop(cat_col, axis=1)
    X_onehot = pd.concat([X, pd.DataFrame(aux_transformed)], axis=1)
    return X_onehot

from sklearn.preprocessing import FunctionTransformer
onehot_pipe  = FunctionTransformer(prepare)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC




clf1 = make_pipeline(onehot_pipe , SVC(gamma='auto', probability=True))
```

```python
clf1.fit(X,y)
```

```python
prepare(df_test)
```

```python
cat_col = df_test.select_dtypes(include=['category']).columns
print(cat_col)
for i in cat_col:
    print(len(df_test[i].unique()))
```

```python

print("Fitting Classifiers")
eclf1 = VotingClassifier(estimators=[
        ('svm', clf1)], voting='hard')





print("Hard Voting ")
hard_scores = cross_val_score(eclf1, X, y, cv=10)

print("aux stop")


# import joblib
# name_of_model = "rf+svm"

# joblib.dump(hard_scores, "Temp/"+name_of_model+"_hard_scores"+".joblib")
# joblib.dump(hard_scores, "Temp/"+ name_of_model+"_soft_scores"+ ".joblib")

print("hard voting", hard_scores.mean())


print("aux stop")
```
