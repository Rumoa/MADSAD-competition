```python
from numpy import zeros
from Script import *
from create_experiments import *
import itertools

algorithms = ['Rf', 'svm', 'xgbhist'] 

columns = ['all', 'no cat']

outliers = ['yes', 'no']

onehot = ['yes', 'no']

include_low_importance_cols = ['yes', 'no']

include_low_unique = ['yes', 'no']


list_of_options = list(itertools.product(algorithms, columns, outliers, onehot, include_low_importance_cols, include_low_unique))

list_of_options = [list(x) for x in list_of_options]
```

```python
from sklearn.inspection import permutation_importance
import time 

from Script import *
df_train = pd.read_csv("Data/train.csv")
df_test  = pd.read_csv("Data/test.csv")

df_train = df_train.drop("i", axis=1)
y = df_train.pop("y")
df_train.insert(54,"y", y)
df_test = df_test.drop(["y", "i"], axis = 1)


df_train, df_test = preprocess_data(df_train, df_test, verbose=False)
```

```python
options = pd.DataFrame(list_of_options)
names = ['algorithms', 'columns', 'outliers', 'onehot', 'include_low_importance_cols', 'include_low_unique']
options.columns = names
options
```

```python
results = []
for i in range(len(list_of_options)):
    print(dict(zip(names,list_of_options[i])))
    out, out_mean = train_with_options(df_train, df_test, list_of_options[i], ncv=3)
    results.append(out_mean)
```

```python
res = {'acc': results}
b = pd.DataFrame(res)
options["acc"] = results
import joblib
joblib.dump(options, "Temp/dataframe_3cv_rf_svm_xgb.joblib")
```
